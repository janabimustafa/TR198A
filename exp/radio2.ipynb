```python
import broadlink.remote as remote
from manchester_code import decode, decode_bits
from collections import Counter
import statistics as st
from typing import List
import base64

def show(bits):
    s = "".join(map(str, bits))
    idx = "".join(f"{i%10}" for i in range(len(bits)))  # 0-9 repeating
    print(s)
    print(idx)

def _threshold(data):
    """Return a good short/long threshold for Broadlink pulses."""
    data = sorted(data)
    if len(data) < 4:                      # too few samples → fallback
        return st.median(data)

    # two -cluster split: lower ⅓ vs upper ⅔
    q1, q2 = st.quantiles(data, n=3)
    return (q1 + q2) / 2                  # mid-point between the clusters
def manchester_decode(durations, gap=2000):
    """Return (bitstring, start_index) from a raw Broadlink pulse list."""
    DATA = [t for t in durations if t < gap]          # 1. cut inter-frame gaps
    if len(DATA) < 4:
        raise ValueError("Not enough data after gap-filtering")

    THR     = _threshold(DATA)                        # 2. quantise
    symbols = ['S' if t < THR else 'L' for t in DATA]

    # 3. auto-resync
    bits, i = [], 0
    while i + 1 < len(symbols):
        a, b = symbols[i], symbols[i + 1]

        if a == b:            # LL or SS → still in leader or a timing glitch
            i += 1            # shift by one symbol and try again
            continue

        bits.append('0' if (a, b) == ('S', 'L') else '1')
        i += 2

    return ''.join(bits), i


def crc8_maxim_reflected(buf: bytes) -> int:
    """Dallas/Maxim CRC-8, LSB-first (poly 0x8C = 0x31 reflected)."""
    crc = 0
    for b in buf:
        crc ^= b
        for _ in range(8):
            crc = (crc >> 1) ^ 0x8C if (crc & 0x01) else (crc >> 1)
    return crc & 0xFF

def decode_56(bits: str) -> dict:
    # bits comes in MSB-first; reverse each byte to get on-air LSB-first order
    bytes_lsb_first = [int(bits[i:i+8][::-1], 2) for i in range(0, 56, 8)]
    id_lo, id_hi, cmd_a, cmd_b, cmd_c, mirror, crc = bytes_lsb_first

    motor   = bool(cmd_a & 0x40)
    payload = (cmd_a >> 2) & 0x0F          # speed or brightness
    light   = bool(cmd_a & 0x80)
    breeze  = bool(cmd_a & 0x02)

    direction = bool(mirror & 0x01)        # 0 = fwd, 1 = rev

    timer_raw = (cmd_c >> 5) & 0x07
    timer_map = {6:0, 5:2, 3:4, 1:8}
    timer = timer_map.get(timer_raw, None)

    mirror_ok = mirror == (((cmd_a >> 1) & 0x3F) | direction)
    crc_ok    = crc == crc8_maxim_reflected(bytes(bytes_lsb_first[:6]))

    return {
        "remote_id": (id_hi << 8) | id_lo,
        "light":     light,
        "motor_on":  motor,
        "speed":     payload if motor else None,
        "brightness": payload if not motor else None,
        "breeze":    breeze,
        "direction": "reverse" if direction else "forward",
        "timer_h":   timer,
        "mirror_ok": mirror_ok,
        "crc_ok":    crc_ok,
        "raw_bits":  bits,
        "tx_id":     f"{id_hi:02X}{id_lo:02X}",
    }
def decode_capture(name, bytes):
    pulses = remote.data_to_pulses(bytes)
    print(f"\n{name} pulses ({len(pulses)}):")
    print("Raw pulses:", "".join(str(b) + ', ' for b in pulses))
    buckets = Counter(round(p/50)*50 for p in pulses if 250 < p < 2_000)
    print("Top pulse buckets (µs,count):", buckets.most_common(4))
    # --- find Manchester short/long widths (ignore gaps >2 ms) --------------
    (short_us, _), (long_us, _) = sorted(
        buckets
        .most_common(2),
        key=lambda kv: kv[0],
    )

    
    thr = (short_us + long_us) / 2
    #skip first two pulses
    # Trim leading gaps > 2 ms -----------------------------------------------
    # print(format_pulses(pulses))
    # pulses = pulses[3:]
    # pulse_bits = [True if p > thr else False for p in pulses]
    # pulse_values = list(decode_bits(pulse_bits))
    # data = []    
    
# # --- nudge alignment if we started mid-mark -------------------------------
#     if (pulses[0] < thr) == (pulses[1] < thr):
#         pulses = pulses[2:]                      # discard first fragment
#     # --- identify start-of-frame: first valid Manchester pair ---------------
    #pulses = pulses[1:]
    data, start = manchester_decode(pulses)
        # if len(data) == 56:
        #     break
    print(f"{name}:  bits={len(data)}   (short≈{short_us} µs, long≈{long_us} µs)")
    print(f"{name} bits:")
    show(data[:56])  # cut to first 56 bits (one frame)

    if len(data) < 56:
        raise ValueError("Frame too short – header not trimmed correctly?")
    

    # ---- cut to first 56 bits (one frame) ----------------------------------
    #0–15 Remote ID Little-endian 16-bit identifier learned by the receiver
    #16–23 CMD A bit 7 = Light toggle; bit 6 = Fan-motor enable; bits 5-2 = Speed 0…9; bit 1 = Breeze/Natural-wind; bit 0 = reserved
    #24–31 CMD B Model / group code (0xCA on all HDC units seen)
    #32–37 CMD C bits 7-5 = Timer (000=none, 001=2 h, 010=4 h, 100=8 h) ; bit 3 = Direction (1 = reverse); bits 2-0 = reserved
    #38–42 Mirror Bit-for-bit copy of bits 32–37 (cheap error check)
    #48–55 CRC-8 Dallas/Maxim poly 0x31, init 0x00, over bytes 0-5


    bits = data[:56]

    return decode_56("".join(map(str, bits)))

decoded = {k: decode_capture(k, v) for k, v in CAPTURES.items()}

# ─── compare the two frames --------------------------------------------------
if decoded["direction_up"]["raw_bits"] == decoded["direction_down"]["raw_bits"]:
    print("\n⚠️  The 56-bit payload is identical – you probably started "
          "decoding in the wrong place or clipped the header differently.")
else:
    print("\n✅  Frames differ as expected:")
    for k in decoded["direction_up"].keys():
        print(f"  {k:>6}: {decoded['direction_up'][k]}  →  {decoded['direction_down'][k]}")
```

```python
import itertools
tick = 32.84                                   # Broadlink time-unit (µs)

def u(n):                                      # n = number of units wanted
    return n * tick + 0.5                      # 0.5 µs pushes us over the edge
def pulses_to_data_round(pulses, tick=32.84):
    out = bytearray(b'\x26\x00\x00\x00')
    for p in pulses:
        u = int(round(p / tick))              # <-- round instead of floor
        if u >= 256:
            out += b'\x00' + u.to_bytes(2, 'big')
        else:
            out.append(u)
    length = len(out) - 4
    out[2:4] = length.to_bytes(2, 'little')
    return out
sample_command = '10101101010010001100000'
# MARK0, SPACE0 = 400, 700
# MARK1, SPACE1 = 700, 400

# send_pulses = [1336916]
# for b in sample_command:
#     if b == "0":
#         send_pulses += [MARK0, SPACE0]
#     else:
#         send_pulses += [MARK1, SPACE1]
# pkt = remote.pulses_to_data(send_pulses)
# pkt[0] = 0xB1
# pkt[1] = 0xC0


# --- 1. choose timings identical to the working remote -------------
T0H, T0L = 394, 755        # 0 bit: 11 u mark, 23 u space
T1H, T1L = 755, 394        # 1 bit: 23 u mark, 11 u space
lead_in_gap     = 1336916               # 1.336 s – wake the receiver
tiny_gap        = 92805
interframe_gap  =    11822               # 50 ms between repeats
trailer_gap     =    49_260               # 49 ms quiet at the very end
repeats         = 4                       # what the OEM remote does

# ---- 2. one frame of mark/space pulses ----------------------------
frame =  list(itertools.chain.from_iterable(
           (T0H, T0L) if b == "0" else (T1H, T1L) for b in sample_command))

# ---- 3. full transmission with lead-in, repeats and trailer -------
send_pulses = [lead_in_gap, tiny_gap, T1H, T1L, T1H] + frame
for _ in range(repeats - 1):
    send_pulses += [interframe_gap, T1L, T1H, T1L, T1L, T1H, T1L, T1H] + frame
send_pulses += [trailer_gap]


# ---- 4. convert µs → Broadlink units, then IR → RF header ----------
packet = pulses_to_data_round(send_pulses)    # first byte is 0x26 (IR)
packet[0] = 0xB1                              # 433 MHz RF mode
packet[1] = 0xC0                              # 32.84 µs time-base
# pulses_to_data already fixed the length bytes (2-3)

working = CAPTURES['direction_up']
working_pulses = remote.data_to_pulses(working)
working_packet = pulses_to_data_round(working_pulses)

print("W  pulses:", "".join(str(755 if b > 500 and b < 1000 else 394 if b < 500 and b > 200 else b) + ', ' for b in working_pulses))
print("My pulses:", "".join(str(b) + ', ' for b in send_pulses))
working_packet[0] = 0xB2
working_packet[1] = 0xC0

print("W  command:", "".join('0x' + format(b, '02x') + ', ' for b in working_packet))
print("My command:", "".join('0x'+format(b, '02x')+', ' for b in packet))
device.send_data(packet) 
```
